## Import the required libraries

```python
# Import the required libraries
import spacy
import string
from collections import Counter
from heapq import nlargest
```

## Open and read the text file

```python
# Open and read the text file
with open('spaceX_DP.txt', 'r') as file:
    text = file.read()
```

```python
# display text in the file
display(text)
```

## Load the SpaCy model and stop words

```python
# Load the SpaCy model and stop words
nlp = spacy.load('en_core_web_sm')
stop_words = spacy.lang.en.STOP_WORDS
```

```python
# Displaying the Stop words
print(stop_words)
```

## Tokenize the 'SpaceX.txt' corpus

```python
# Tokenize the text
doc = nlp(text)
```

```python
# Create a list of words without stop words
words_without_stop = [token.text for token in doc if not token.is_stop]

# Join the words back into a single string
cleaned_text = ' '.join(words_without_stop)

display(cleaned_text)
```

## Build word frequency dictionary after removing stopwords

```python
# Build word frequency dictionary after removing stopwords
word_frequencies = Counter()
for word in doc:
    if word.text.lower() not in stop_words and word.text not in string.punctuation:
        word_frequencies[word.text] += 1

# display(word_frequencies)
```

## Determine the maximum frequency and normalize word frequencies

```python
# Determine the maximum frequency
maximum_frequency = max(word_frequencies.values())

# Normalize word frequencies using the maximum frequency
for word in word_frequencies:
    word_frequencies[word] = word_frequencies[word] / maximum_frequency

# display(word_frequencies)
```

## Tokenize the sentences. Generate the sentence_scores. Score every sentence based on number of words.

```python
# Tokenize sentences and score them based on the number of words
sentences = [sent for sent in doc.sents]
sentence_scores = {sent: sum(word_frequencies[word.text] for word in sent if word.text in word_frequencies) for sent in sentences}
```

## Import nlargest from heapq and provide 'summarized_sentences'.

```python
# using nlargest from heapq and get summarized sentences
num_sentences = 5  #  the number of sentences want in the summary (5)
summarized_sentences = nlargest(num_sentences, sentence_scores, key=sentence_scores.get)
display(summarized_sentences)
```

## Convert sentences from spacy to strings and join all sentences.

```python
# Convert SpaCy sentences to strings and join them for the summary
summary = ' '.join([str(sent) for sent in summarized_sentences])
```

## Provide a summary once you have converted spacy outputs to strings.

```python
# Print the summary
print(summary)
```

## Determine the length of the summary.

```python
# Determine the length of the summary
summary_length = len(summary)
print("summary_length :")
print(summary_length)
```

## Use SpaCy and Textrank to summarize the text

**Note :** Used **Gensim's** Textrank

```python
import spacy
from summa import summarizer

# Load the English NLP model from spaCy
nlp = spacy.load("en_core_web_sm")
```

```python
# Process the text using spaCy
doc = nlp(text)

# Convert the spaCy doc to plain text
plain_text = ' '.join([token.text for token in doc])

# Summarize the text using Gensim's TextRank-based summarization
summary = summarizer.summarize(plain_text, ratio=0.2)
```

## Print the Textrank summary

```python
# Split the summary into sentences
summary_sentences = summary.split(". ")[:5]  # Extract the first 5 sentences

# Print the 5-sentence summary
print("Summary : ")
display('. '.join(summary_sentences))
```
